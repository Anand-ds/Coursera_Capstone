Question 1: Use the BeautifulSoup package or any other way you are comfortable with to transform the data in the table on the Wikipedia page into the above pandas dataframe

Now Importing List of Postal codes Canada from Wiki

```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

Observing output we see that "Wikitable sortable" has the dataset in table format, hence extracting table

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

Printing headers

```python
print(My_table.tr.text)
```

```python
headers="Postcode,Borough,Neighbourhood"
```

fetch all values in tr and separate td by ,

```python
table1=""
for tr in My_table.find_all('tr'):
    row1=""
    for tds in tr.find_all('td'):
        row1=row1+","+tds.text
    table1=table1+row1[1:]
print(table1)
```

Create the toronto csv file with data 

```python
file=open("toronto.csv","wb")
#file.write(bytes(headers,encoding="ascii",errors="ignore"))
file.write(bytes(table1,encoding="ascii",errors="ignore"))
```

Convert this now into a pandas dataframe importing pandas as pd by reading toronto.csv file and assign column names to the information gathered

```python
import pandas as pd
df = pd.read_csv('toronto.csv',header=None)
df.columns=["Postalcode","Borough","Neighbourhood"]
```

Print first 10 lines

```python
df.head(10)
```

Process the data with cells having a designated borough. Drop cells with a not assigned cells

```python
indexNames = df[ df['Borough'] =='Not assigned'].index
df.drop(indexNames , inplace=True)
```

Display the correct designated boroughs now

```python
df.head(10)
```

If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python
df.loc[df['Neighbourhood'] =='Not assigned' , 'Neighbourhood'] = df['Borough']
df.head(10)
```

ALL Rows with similar postalcodes is now combined into one row. 
The neighborhoods separated with a comma

```python
result = df.groupby(['Postalcode','Borough'], sort=False).agg( ', '.join)
```

```python
df_new=result.reset_index()
df_new.head(15)
```

```python
# defining radius and limit of venues to get
radius=500
LIMIT=100
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=Toronto_df['Neighbourhood'],
                                   latitudes=Toronto_df['Latitude'],
                                   longitudes=Toronto_df['Longitude']
                                  )
```
